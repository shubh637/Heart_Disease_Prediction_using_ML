```python
#importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import sklearn as sc
import seaborn as sns
```

```python
df=pd.read_csv("./framingham.csv")
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
columns_to_impute = ['education', 'cigsPerDay', 'BPMeds', 'totChol', 'BMI', 'heartRate', 'glucose']
for col in columns_to_impute:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df[df.columns].corr(),annot=True,fmt=".2f",cmap='coolwarm')
```

```python
#histogrames
fig,axes=plt.subplots(2,8,figsize=(30,20))
df.hist(ax=axes,bins=30,grid=False)
plt.show()
```

```python
#boxplots
plt.figure(figsize=(30,20))
for i in enumerate(df.columns):
    plt.subplot(6,4,i[0]+1)
    sns.boxplot(x=i[1],data=df)
plt.tight_layout()    
```

```python
plt.figure(figsize=(30,20))
plt.boxplot(df)
```

```python
x=df.iloc[:,0:15]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=0)
x_train
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

# Assuming x_train, y_train, x_test, and y_test are already define
# Initialize the Logistic Regression classifier with increased max_iter
classifier = LogisticRegression(max_iter=200)  # You can adjust max_iter as needed

# Fit the model
classifier.fit(x_train, y_train)

# Predict the test set results
y_pred = classifier.predict(x_test)

# Print the accuracy
print("The logistic regression accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm/np.sum(cm),fmt=".2%",annot=True)

```

```python
#classification report
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
